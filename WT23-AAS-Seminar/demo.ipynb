```python
# Get the first 5 single-objective noiseless BBOB function in dimension 2 and 3 for the first two instances.
suite = cocoex.Suite("bbob", f"instances:1-2", f"function_indices:1-5 dimensions:2,3")
print(suite)
suite[0]([1,2])
x = [1,2, 1]
BBOB1 = suite[10]
print(BBOB1)
BBOB1(x)
```

```python
features = []
for problem in suite:
    dim = problem.dimension
    fid = problem.id_function
    iid = problem.id_instance

    # Create sample
    X = create_initial_sample(dim, lower_bound = -5, upper_bound = 5)
    y = X.apply(lambda x: problem(x), axis = 1)

    # Calculate ELA features
    # ela_meta = calculate_ela_meta(X, y)
    ela_distr = calculate_ela_distribution(X, y)
    nbc = calculate_nbc(X, y)
    # disp = calculate_dispersion(X, y)
    # ic = calculate_information_content(X, y, seed = 100)

    # Store results in pandas dataframe
    # data = pd.DataFrame({**{'fid': fid}, **{'dim': dim}, **{'iid': iid}, **ic, **ela_meta, **ela_distr, **nbc, **disp}, index = [0])
    data = pd.DataFrame({**{'fid': fid}, **{'dim': dim}, **{'iid': iid},  **ela_distr, **nbc}, index = [0])
    features.append(data)

features = pd.concat(features).reset_index(drop = True)
```
